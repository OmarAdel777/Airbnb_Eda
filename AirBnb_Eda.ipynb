# Data Exploration and Cleaning
#### We'll begin by exploring the dataset, checking for missing values, duplicates, and unique values in various columns. We'll also clean the data by dropping unnecessary columns and filling missing values.

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
# Display the last few rows of the DataFrame
df.tail()
```

```python
# Get the dimensions of the DataFrame
df.shape
```

```python
# Display information about the DataFrame's columns and data types
df.info()
```

```python
# Check for missing values in each column
df.isna().sum()
```

```python
# Check for duplicated rows
df.duplicated().sum()
```

```python
# Get the number of unique values in each column
df.nunique()
```

```python
# Display unique host names
df.host_name.unique()
```

```python
# Check for duplicated host IDs
df.host_id.duplicated().sum()
```

```python
# Drop columns with excessive missing values or duplicates
df.drop(["neighbourhood_group", "last_review", "license", "host_name", "host_id"], axis=1, inplace=True)
```

```python
# Fill missing values in the 'reviews_per_month' column with zeros

df.reviews_per_month.isnull().sum()
```

```python

# Fill missing values in the 'reviews_per_month' column with zeros
df.fillna({"reviews_per_month": 0}, inplace=True)
df.reviews_per_month.isnull().sum()
```

```python
# Check for any remaining missing values
df.isnull().sum()
```

```python
# Display summary statistics for numeric columns
df.describe().T
```

```python
# Display again information about the DataFrame's columns and data types
df.info()
```

```python
# Convert 'neighbourhood' and 'room_type' columns to categorical for efficiency
df["neighbourhood"] = df["neighbourhood"].astype("category")
df["room_type"] = df["room_type"].astype("category")
```

```python
# Clean column names by replacing underscores and converting to title case
df.columns = df.columns.str.replace('_', ' ').str.title()
```

```python
# Replace a specific text format in the 'Room Type' column
df['Room Type'].replace('Entire home/apt', 'Entire Home/Apt', inplace=True)
```

```python
# Rename the 'Price' column to 'Price in Dollars'
df.rename(columns={'Price': 'Price in Dollars'}, inplace=True)
```

```python
# Display counts of listings per neighborhood
df['Neighbourhood'].value_counts()
```

# Correlation Heatmap:The heatmap provides an overview of correlations between numeric variables. 



```python
# Calculate the correlation matrix and visualize it as a heatmap
correlation = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.show()
```

# Data Insights and Visualizations
## Price Distribution
### The histogram of the 'Price in Dollars' column provides an overview of the distribution of listing prices. This helps us understand the range of prices and the most common price ranges.

```python
# Create a histogram of the 'Price in Dollars' column
plt.figure(figsize=(12, 6))
sns.histplot(data=df["Price in Dollars"], bins=10, kde=False , color='blue')
plt.xlabel('Price in Dollar')
plt.ylabel('Frequency')
plt.title('Histogram of Price in Dollar')
plt.show()
```

# Aggregated Price Analysis:
## The aggregated data by 'Neighbourhood', 'Room Type', and 'Availability 365' reveals that some neighbourhoods and room types consistently have higher average prices throughout the year.

```python
# Group and aggregate data to calculate mean prices by 'Neighbourhood', 'Room Type', and 'Availability 365'
n = df.groupby(['Neighbourhood', 'Room Type', 'Availability 365'])['Price in Dollars'].mean().sort_values(ascending=False).round(1)
n
```

```python
# Reshape the aggregated data for better visualization
n1 = n.unstack(level=1)
n1
```

```python
# Group and aggregate data 
neighbour_price = df.groupby(['Neighbourhood'])['Price in Dollars'].mean().sort_values(ascending=False).round(1)
neighbour_price
```

```python
# Plot the data
plt.figure(figsize=(15, 8))
neighbour_price.plot(kind='bar', stacked=True)
plt.title('Average Price of Room Types in Each Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price in Dollars')
plt.legend(title='Room Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
Room_price = df.groupby(['Room Type'])['Price in Dollars'].mean().sort_values(ascending=False).round(1)
Room_price
```

```python
# Group and aggregate data
Room_price = df.groupby(['Room Type'])['Price in Dollars'].mean().sort_values(ascending=False).round(1)

# Plot the data
plt.figure(figsize=(8, 6))
Room_price.plot(kind='bar', color='blue', alpha=0.7)
plt.title('Average Price of Each Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price in Dollars')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

# Plot the average price of room types in each neighbourhood
neighbour_price.plot(kind='bar', stacked=True, ax=axes[0])
axes[0].set_title('Average Price of Room Types in Each Neighbourhood')
axes[0].set_xlabel('Neighbourhood')
axes[0].set_ylabel('Average Price in Dollars')
axes[0].legend(title='Room Type')
axes[0].tick_params(axis='x', rotation=90)

# Plot the average price of each room type
Room_price.plot(kind='bar', color='blue', alpha=0.7, ax=axes[1])
axes[1].set_title('Average Price of Each Room Type')
axes[1].set_xlabel('Room Type')
axes[1].set_ylabel('Average Price in Dollars')
axes[1].tick_params(axis='x', rotation=45)

# Adjust the layout of the subplots
plt.tight_layout()
plt.show()
```
